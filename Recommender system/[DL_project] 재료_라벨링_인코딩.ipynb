```python
def filter_ingredients_in_class(ingredient_list):
    return [ingredient for ingredient in ingredient_list if any(robo in ingredient for robo in roboflow_class)]

df_filter_ingredients['class_ingredient'] = df2['new_ingredients'].apply(filter_ingredients_in_class)
```

##  One-hot Encoding

### 레시피별 식재료 [0,1,1,0,0,0 ...] 이런식으로 표현되게끔

```python
len(robo_ing_dict)
```

```python
## One-Hot Encoding

total_ingredients = 83

def one_hot_encoding(row, total_ingredients): 
    encoded_ingredients = np.zeros(total_ingredients, dtype=int) # 클래스 총 길이만큼 값을 0으로 초기화
    for ingredient in row:
        encoded_ingredients[ingredient] = 1
    return encoded_ingredients
```

```python
import numpy as np

only_robo_df['emb_num'] = only_robo_df['new_ingredients_idx'].apply(lambda x: one_hot_encoding(x, total_ingredients))
```
